<div><h1>Difference Between Concurrency and Parallelism</h1><br><img src="images/photo_2025-03-03_20-07-46.jpg"></img><br></div>

<div><p>In concurrency, tasks appear to run at the same time. In reality, however, only one task is executed at any given moment, and the system rapidly switches between tasks so that they seem to be running simultaneously. This is called a context switch. In parallelism, tasks are actually executed at the same time by using two or more CPU cores.</p><img src="images/photo_2025-03-03_20-12-04.jpg"></img></div>

<div><h4>Real-World Example of Concurrency:</h4><p>Consider a chef who is chopping, stirring, adjusting the heat, mixing ingredients, and performing other activities seemingly at the same time. In reality, the chef is rapidly switching between these tasks.</p><img src="images/2025-07-11_15-19-03.png"></img><h4>Real-World Example of Parallelism:</h4><p>Consider two or more chefs, each responsible for a specific task. Here, the tasks are truly being performed simultaneously.</p><img src="images/2025-07-11_15-23-55.png"></img></div>

<div><h2>Important Concepts in Python Concurrency</h2><h4>What is an Execution Unit?</h4><p>In concurrency, an execution unit is a component or entity capable of executing and managing one or more tasks concurrently. Multiple execution units can also work together to perform several tasks at the same time.</p><h5>Execution Units in Python:</h5><p>1) Process<br>2) Thread<br>3) Coroutine</p><p><b>What is a Process?</b> A process is an instance of a running computer program that occupies part of the system resources. Each process can run independently and has its own dedicated resources.</p><p><b>What is a Thread?</b> A thread is the smallest unit of execution that can run in a system. A thread consists of a sequence of instructions and runs within a process. However, it is not correct to define a process simply as a collection of threads. For example, Word and Excel are two separate processes. Word contains multiple threads for tasks such as copying, selecting, text alignment, and more. The operating system is responsible for managing these threads.</p><p><b>What is a Coroutine?</b> A coroutine is an execution unit that can suspend itself and resume later. In Python, coroutines are objects containing a set of instructions that can be executed whenever needed. Coroutines are managed by the program itself and enable concurrency. They run inside threads, but it is not correct to say that threads are made up of coroutines.</p><img src="images/2025-07-11_15-49-55.png"></img></div>
